```python
#importing python libraries to be utilized
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from os.path import join
```

```python
#opening the csv files for each dataset to be utilized
country = pd.read_csv("/Users/hayrettinsendag/Documents/GitHub/EduRank/datasets/country.csv", encoding='UTF-8')
enrollment = pd.read_csv("/Users/hayrettinsendag/Documents/GitHub/EduRank/datasets/enrollment.csv", delimiter=';', quotechar='"', names=['enr_id', 'year', 'country_code', 'ratio'], header=None)
expenditures = pd.read_csv("/Users/hayrettinsendag/Documents/GitHub/EduRank/datasets/expenditures.csv", delimiter=';', quotechar='"', names=['exp_id', 'year', 'exp_percentage', 'country_code'], header=None)
literacy_reports = pd.read_csv("/Users/hayrettinsendag/Documents/GitHub/EduRank/datasets/literacy_reports.csv", encoding='UTF-8')
teacher_quality = pd.read_csv("/Users/hayrettinsendag/Documents/GitHub/EduRank/datasets/teacher_quality.csv", delimiter=';', quotechar='"', names=['quality_id', 'country_code', 'percentage', 'year'], header=None)
```

```python


# Create a distribution of literate ages for all world
min = literacy_reports['year'].min()
max = literacy_reports['year'].max()
med = literacy_reports['year'].median()
timeframe = max - min
#med_year = literacy_expenditure['year'].mean()
literacy_reports = literacy_reports.drop_duplicates()
wanted_year = literacy_reports[literacy_reports['year'] > med ]
print("The histogram below is valid for years between ",min,"-",med)
plt.hist(wanted_year['age'])
plt.xlabel('age')
plt.ylabel('rate')
plt.title('Distribution of Literate Ages')
plt.show()
```

```python

teacher_enrollment = pd.merge(teacher_quality,enrollment,on='country_code')
wanted = teacher_enrollment[teacher_enrollment['ratio']< 90]
plt.hist(wanted['ratio'])
plt.ylabel('Enrollment Ratio')
plt.xlabel(' Teacher Quality Percentage')
plt.title('Teacher quality efect on enrollment Ratio ')
plt.show()

```

```python
country_literacy = pd.merge(literacy_reports,country,on='country_code')
country_literacy = country_literacy.drop_duplicates
rate_mean = country_literacy['rate'].mean()
rate_std = country_literacy['rate'].std()
bad_rate = country_literacy[country_literacy['rate'] < rate_mean-rate_std]

# Plotting a bar chart of country codes
plt.figure(figsize=(10, 6))
plt.bar(bad_rate['country_code'], bad_rate['rate'])
plt.xlabel('Country Code')
plt.ylabel('Literacy Ratio')
plt.title('Countries that are far behind ')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
